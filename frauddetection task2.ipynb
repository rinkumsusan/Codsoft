```python
train_df['trans_date_trans_time'] = pd.to_datetime(train_df['trans_date_trans_time'])

# Convert 'dob' to datetime
train_df['dob'] = pd.to_datetime(train_df['dob'])

# Extract useful features from transaction date
train_df['Transaction_Year'] = train_df['trans_date_trans_time'].dt.year
train_df['Transaction_Month'] = train_df['trans_date_trans_time'].dt.month
train_df['Transaction_Day'] = train_df['trans_date_trans_time'].dt.day
train_df['Transaction_Hour'] = train_df['trans_date_trans_time'].dt.hour
train_df['Transaction_Weekday'] = train_df['trans_date_trans_time'].dt.weekday  # 0 = Monday, 6 = Sunday

# Convert 'dob' to 'Age' (More useful than raw DOB)
train_df['Age'] =train_df['dob'].apply(lambda x: 2025 - x.year)  # Assuming analysis is in 2025

# Drop original datetime columns (optional, if not needed)
train_df = train_df.drop(columns=['trans_date_trans_time', 'dob'])

# Save preprocessed dataset (optional)
train_df.to_csv("preprocessed_data.csv", index=False)
```

```python
x_train = train_df.drop(columns=['is_fraud'])  # Drop target column
y_train = train_df['is_fraud']
```

```python
print(x_train.dtypes)  # Check data types
print(x_train.head())  # View first few rows
```

```python
x_train = x_train.drop(columns=['merchant','unix_time','Unnamed: 0','cc_num','merchant','first','last','street','zip','trans_num','city'])



```

```python
x_train.gender=x_train.gender.apply(lambda x: 1 if x=="M" else 0)
x_train.gender.value_counts()
x_train.head()
```

```python
x_train = x_train.drop(columns=['job','merch_lat','merch_long','long','lat'])

#for col in train_df.select_dtypes(include=['object']).columns:
#    train_df[col] = train_df[col].astype('category')
```

```python
x_train = pd.get_dummies(x_train, columns=['category'], drop_first=True)

#x_train = x_train.astype({col: 'float32' for col in df.select_dtypes(include=['float64']).columns})

```

```python
x_train.head()
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
print("Scaling completed successfully!")
```

```python
model = LogisticRegression()
model.fit(x_train_scaled, y_train)
```

```python
y_train_pred = model.predict(x_train_scaled)
```

```python
train_accuracy = accuracy_score(y_train, y_train_pred)
print(f"Training Accuracy: {train_accuracy:.2f}")
print("\nClassification Report (Training Data):\n", classification_report(y_train, y_train_pred))
print("\nConfusion Matrix (Training Data):\n", confusion_matrix(y_train, y_train_pred))
```
